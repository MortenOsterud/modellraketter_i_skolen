## Beregninger i oppstartsfasen – Sannsynliggjøring av at prosjektet kan gjennomføres

Før man kan skyte opp raketter i skolen er det mange faser av
sikkerhetsvurderinger som må gjennomføres. Hvor mye tid og ressurser
disse beregningene tar varierer. Derfor er det alltid hensiktsmessig å
gjøre en første vurdering av hvorvidt prosjektet kan være
gjennomførbart. Enkle fysikkberegninger kan hjelpe deg med å avgjøre om
en rakett vil forlate launchpad, og hvor høyt den maksimalt kan gå.

La oss ta et ganske vanlig skoleekspempel: En liten modellrakett med en
C eller D-klasse-motor. Før vi legger ned mangfoldige timer i design,
bygging, forberedende pedagogisk arbeid etc må vi forsikre oss om at
prosjektet er gjennomførbart. Ifølge dronekommandoen til politiet i Oslo
må man ha følgende kriterier på plass om man ikke er sertifisert
dronepilot og skal skyte opp en modellrakett i undervisningssammenheng
uten spesielle tillatelser.

**Sikkerhetskriterier:**  
Rakettmotorklasse: D eller lavere (20Ns eller mindre)  
Totalvekt: Under 250g  
Maksimal flyhøyde: 120 meter  
Annet: Sikkerhetssone, tillatelse fra grunneier, risikovurdering etc

I første fase av sikkerhetsberegninger bruker jeg de fire
bevegelseslikningene fra fysikk1 ettersom de fleste realfagslærere er
kjent med disse:

**Bevegelseslikninger**

$a = akselerasjon\ \left( - 9,81m{/s}^{2}\ \ på\ jorda \right)$  
$s = strekning\ (meter)$ *  
*$t = tid\ (sekunder)$ *  
*$v_{0} = \ startfart\ (m/s)$ *  
*$v = \ sluttfart\ (m/s)$

$I:\ \ \ \ \ \ \ \ \ \ \ \ \ v = at + v_{0}$

$II:\ \ \ \ \ \ \ \ \ \ \ \ s = \frac{1}{2}\left( v + v_{0} \right)t$

$III:\ \ \ \ \ 2as = v^{2} - v_{0}^{2}$

$IV:\ \ \ \ \ \ \ \ \ \ \ s = v_{0}t + \frac{1}{2}at^{2}$

Når vi skal gjøre enkle fysikkberegninger har vi naturligvis mulighet
til å gjøre disse for hånd, men her kan digitale verktøy spare både tid,
og gjøre utregningene mer oversiktlige for oss. Dessuten kan de gjøre
beregningene mer tilgjengelige for elever, da de kan følge utregningene
stegvis og kvalitetssikre at koden faktisk regner ut det man tror. Her
er to svært grove beregninger jeg gjorde med en masse på 0.249kg og de
to motor-typene jeg har tilgjengelige: C6 og D9.

Kjør koden nedenfor ved å plassere markøren i cellen og trykke shift+enter


```python
total_masse = 0.249    # kg
motor = 'D9'           # Enten 'D9', 'C6' eller 'B4'

def bestem_kraft_og_tid(motor):
    """Setter korrekt motorkraft og brenntid basert på motortype"""
    if motor == 'D9':
        motor_kraft = 9        # N (kgms^-2)
        motor_tid = 2.1        # s
    elif motor == 'C6':
        motor_kraft = 6        # N (kgms^-2)
        motor_tid = 1.6        # s
    elif motor == 'B4':
        motor_kraft = 4        # N (kgms^-2)
        motor_tid = 1.3        # s
    else:
        motor_kraft = 0
        motor_tid = 0
        
    return motor_kraft, motor_tid

def finn_oppskytningsakselerasjon(masse, kraft, g = 9.81):
    """Beregner konstant akselerasjonen for rakett skutt rett opp gitt motorkraft"""
    akselerasjon = kraft/masse - g
    return akselerasjon

def finn_sluttfart(a, t, v_0 = 0):
    """Bevegelseslikning 1. Finner sluttfart gitt akselerasjon, tid og evt startfart"""
    v = a*t + v_0
    return v

def finn_strekning(v, t, v_0 = 0):
    """Bevegelseslikning 2. Finner strekning gitt startfart, sluttfart (valgfri) og tid"""
    s = 0.5*(v+v_0)*t
    return s

def finn_strekning2(v, a, v_0 = 0):
    """Bevegelseslikning 3, omskrevet med hensyn på s. Finn strekning til apogee etter motoravstengning"""
    s = (v**2-v_0**2)/(2*a)
    return s

motor_kraft, motor_tid = bestem_kraft_og_tid(motor)
oppskytningsakselerasjon = finn_oppskytningsakselerasjon(total_masse, motor_kraft)
v_motorstopp = finn_sluttfart(oppskytningsakselerasjon, motor_tid)
h_motorstopp = finn_strekning(v_motorstopp, motor_tid)
h_coasting = finn_strekning2(v_motorstopp, g)
h_total = h_motorstopp+h_coasting

print(f"Akselerasjon: {round(oppskytningsakselerasjon,3)} m/s^2")
print(f"Maksfart: {round(v_motorstopp,2)} m/s")
print(f"Høyde MECO: {round(h_motorstopp,1)} m")
print(f"Høyde coast: {round(h_coasting,1)} m")
print(f"Total høyde u/luftmotstand: {round(h_total,1)}m")
```

sammen med matplotlib til å vise hvordan variasjoner i masse påvirker
forventet høyde. Her er en revidert kode som bruker matplotlib til å vise forventet høyde med ulike masser med en gitt motor. De ulike beregningene har også blitt samlet i én funksjon.

## Beregninger i designfasen

I designfasen av rakettprosjektet kan man velge å gå en relativt trygg
vei ved å bruke ferdige rakettsett fra f.eks Klima, følge instruksjoner
fra eksempelvis Andøya Space Centers strålende hefte *Modellraketter,
Aktiviteter for fysikken i fokus* (vedlagt etter tillatelse fra Andøya
Space Center) eller man kan designe raketter fra grunnen selv. Uavhengig
av hvordan man går frem her er det helt nødvendig å gjøre sine egne
sikkerhetsberegninger og simuleringer før oppskyting. Det er opp til

## Andre beregninger

Andre ledd av sikkerhetsberegningene kan brukes til å kvalitetssikre
simuleringer fra OpenRocket. OpenRocket er svært avansert og pålitelig
programvare, men vi kan ha vært uheldige under designet, eller ha
aktivert funksjoner vi ikke har oversikt over. Derfor er det anbefalt å
ha et ekstra sikkerhetsnett. Jeg anbefaler enten å utvikle selv, eller
benytte seg av en numerisk baneberegning som hensyntar luftmotstand.
Vedlagt er et strålende python-program utviklet av Fredrik B. Pettersen
hos Andøya Space Education ved navn *NumeriskRakett_eksempel.py*. Det
anbefales å bruke versjonen uten ferdigutfylte variabler
(*NumeriskRakett_mal.py)* slik at du og elevene manuelt må fylle inn
alle variabler, både for maksimalt læringsutbytte og for å sikre at det
brukes korrekte variabler.

```python
# -*- coding: utf-8 -*-
"""
Dette koden er rammeverket for en enkel numerisk beregning av rakettbaner. I denne koden
er x retning horisonalt, mens y retning vertikalt. For at koden skal fungere må følgende oppgaver løses: 
    - konstanter for utskytingsparametre, rakettmotor, vekt og aerodynamikk må fylles inn
    - uttrykk for baneberegninger i while-løkkene må skrives inn. Her skal Newtons 2 lov brukes til å finne akselerasjon, og 
    bevegelsesligningene for å finne hastighet og posisjon

Andøya Space Education

Opprettet: 11.10.2023
Sist endret: 11.10.2023
@forfatter: Fredrik B. Pettersen
"""

# =========================================================================== #
# ============================== Blblioteker ================================ #
import numpy as np # for matematiske beregninger
import matplotlib.pyplot as plt # for plotting

# =========================================================================== #
# =============================== Konstanter ================================ #
# UTSKYTINGSPARAMETRE (SKRIV INN VERDIER!)
theta = 80 # planlagt elevasjon (utskytningsvinkel) målt i grader
g = 9.81 # standard tyngdeakselerasjon i m/s^2

# RAKETTMOTOR (SKRIV INN VERDIER!)
T_avg = 6.765 # gjennomsnittelig skyvekraft i Newton
t_b = 1.565 # motorens brenntid i sekunder
t_d = 3 # motorens glidetid i sekunder

# VEKT (SKRIV INN VERDIER!)
m_0 = 0.145 # rakettens totale startmasse inkludert motor med drivstoff i kg
m_p = 0.0095 # drivstoffmassen i kg
m_avg = m_0 + m_p/2 # gjennomsnittelig masse i kg

# AERODYNAMIKK (SKRIV INN VERDIER!)
C_D = 0.81 # rakettens luftmotstandskoeffisient
rho = 1.225 # lufttetthet i kg/m^3 
A = np.pi*0.0185**2 + 4*(0.048*0.0015 + 0.002*0.0065) # rakettens største tverrsnittareal sett fra fronten i m^2
C_D_P = 1.75 # fallskjermens luftmotstandskoeffisient
A_P = np.pi*0.1**2 # fallskjermens areal i m^2
# print(A)

# SIMULERINGSKONSTANTER
dt = 0.01 # tidssteg mellom hver beregning målt i sekund
t_f = 100 # total simuleringstid mål i sekund

# =========================================================================== #
# =================== Initisalisere numeriske beregninger =================== #

# Elevasjonen regnes om til radianer
theta = theta*np.pi/180 

# Alle verdier som skal simuleres må ha en liste hvor hver enkelt beregning kan legges inn
t = np.arange(t_f, step = dt) # tid                         (starter på 0 s)
a_x = np.zeros(np.shape(t)) # horisontal akselerasjon       (starter med 0 m/s^2)
a_y = np.zeros(np.shape(t)) # vertikal akselerasjon         (starter med 0 m/s^2)
v_x = np.zeros(np.shape(t)) # horisontal hastighet          (starter med 0 m/s)
v_y = np.zeros(np.shape(t)) # vertikal hastighet            (starter med 0 m/s)
x = np.zeros(np.shape(t)) # lengde                          (starter fra 0 m)
y = np.zeros(np.shape(t)) # høyde                           (starter fra 0 m)
m = np.full(np.shape(t), m_0) # masse                       (starter med startmassen i kg)

# definerer en indeks (i) som holder styr på hvor vi er i listene våre
i = 0 # starter på 0
i_max = np.size(t)-1 # maksimal indeks i listene våre; hvis i blir større er vi utenfor listene


# konstantfaktorer i beregningene KAN defineres utenfor løkkene for å spare tid
beta = 0.5*C_D*rho*A # konstant i uttrykket for rakettens luftmotstand
beta_P = 0.5*C_D_P*rho*A_P # konstant i uttrykket for fallskjermens luftmotstand 
T_x = T_avg*np.cos(theta) # konstant skyvekraft fra motoren i x-retning
T_y = T_avg*np.sin(theta) # konstant skyvekraft fra motoren i y-retning


# =========================================================================== #
# =========================== Numeriske beregninger ========================= #

while t[i] < t_b: # Baneberegninger mens motoren brenner, dvs. når t er mindre enn brenntiden til motoren
    # Fyll inn utrykk for akselerasjon i x- og y-retning (Newtons 2 lov) 
    a_x[i+1] = 1/m_avg*(T_x - beta*v_x[i]*np.sqrt(v_x[i]**2 + v_y[i]**2)) 
    a_y[i+1] = 1/m_avg*(T_y - beta*v_y[i]*np.sqrt(v_x[i]**2 + v_y[i]**2)) - g 
        
    # Hastighet i x- og y-retning. Bruk bevegelsesligningen for hastighet
    v_x[i+1] = a_x[i]*dt + v_x[i]
    v_y[i+1] = a_y[i]*dt + v_y[i]
    
    # Posisjon i x- og y retning. Bruk bevegelsesligningen for posisjon
    x[i+1] = 0.5*a_x[i]*dt**2 + v_x[i]*dt + x[i]
    y[i+1] = 0.5*a_y[i]*dt**2 + v_y[i]*dt + y[i]
    
    # Oppdater i. Hvor hver beregning øker indeksen med 1
    i += 1

while t[i] <= t_b + t_d: # Baneberegninger når raketten er i glideflukt. Dvs. motoren har sluttet å brenne, men fallskjermen har enda ikke kommet ut
    # Akselerasjon 
    a_x[i+1] = 1/m_avg*(- beta*v_x[i]*np.sqrt(v_x[i]**2 + v_y[i]**2)) 
    a_y[i+1] = 1/m_avg*(- beta*v_y[i]*np.sqrt(v_x[i]**2 + v_y[i]**2)) - g 
        
    # Hastighet
    v_x[i+1] = a_x[i]*dt + v_x[i]
    v_y[i+1] = a_y[i]*dt + v_y[i]
    
    # Posisjon
    x[i+1] = 0.5*a_x[i]*dt**2 + v_x[i]*dt + x[i]
    y[i+1] = 0.5*a_y[i]*dt**2 + v_y[i]*dt + y[i]
    
    # Oppdater i
    i += 1

while i < i_max and y[i-1] >= 0: # etter at fallskjermen har kommet ut. Dvs. når tiden som har gått er lengre enn brenntid + glidetid   
    # Aksekerasjon
    a_x[i+1] = 1/m_avg*(- beta_P*v_x[i]*np.sqrt(v_x[i]**2 + v_y[i]**2)) 
    a_y[i+1] = 1/m_avg*(- beta_P*v_y[i]*np.sqrt(v_x[i]**2 + v_y[i]**2)) - g 
        
    # Hastighet
    v_x[i+1] = a_x[i]*dt + v_x[i]
    v_y[i+1] = a_y[i]*dt + v_y[i]
    
    # Posisjon
    x[i+1] = 0.5*a_x[i]*dt**2 + v_x[i]*dt + x[i]
    y[i+1] = 0.5*a_y[i]*dt**2 + v_y[i]*dt + y[i]
    
    # Oppdater i
    i += 1

# Raketten har krasjet, så vi sletter unødvendige data
t = t[:i]
x = x[:i]
y = y[:i]
v_x = v_x[:i]
v_y = v_y[:i]
a_x = a_x[:i]
a_y = a_y[:i]
m = m[:i]

# =========================================================================== #
# ================================ Plotting ================================= #

# Først lukker vi alle åpne figurer, slik at vi slipper å gjøre det manuelt
plt.close('all')

# Høyde over lengde
plt.figure('Høyde over avstand')
plt.plot(x,y)
plt.xlabel("Lengde [m]")
plt.ylabel("Høyde [m]")
plt.gca().set_aspect(1) # Sørger for at forholdet mellom aksene er 1:1.
plt.grid(linestyle='--')
plt.show()

# Høyde over tid
plt.figure('Høyde over tid')
plt.plot(t, y)
plt.xlabel("Tid [s]")
plt.ylabel("Høyde [m]")
plt.grid(linestyle='--')
plt.show()

# Hastighet over tid
plt.figure('Hastighet over tid')
plt.plot(t, v_x, label='$v_x$')
plt.plot(t, v_y, label='$v_y$')
plt.plot(t, np.sqrt(v_x**2 + v_y**2), label=r'$|\vec{v}|$')
plt.xlabel("Tid [s]")
plt.ylabel("Fart [m/s]")
plt.grid(linestyle='--')
plt.legend()
plt.show()

# Akselerasjon over tid
plt.figure('Akselerasjon over tid')
plt.plot(t, a_x, label='$a_x$')
plt.plot(t, a_y, label='$a_y$')
plt.plot(t, np.sqrt(a_x**2 + a_y**2), label=r'$|\vec{a}|$')
plt.xlabel("Tid [s]")
plt.ylabel("Akselerasjon [$m/s^2$]")
plt.grid(linestyle='--')
plt.legend()
plt.show()
```

sjekke at dette ligger minst et kaliber (en diameter av raketten) foran
center of pressure.

Center of pressure kan beregnes ved en såkalt skyggemodell. [1]Da tegnes
skyggen til raketten på et ark og du finner det geometriske
gjennomsnittet av skyggefiguren. Hvis man kutter ut skyggen i papp eller
plast kan man også finne vippepunktet til skyggefiguren.

Jeg vil også anbefale å sjekke stabilitet ved snor-fremgangsmåten[2]. Da
fester man en snor rundt raketten og snurrer den rundt seg for å se om
den flyr stabilt. Begynn ved center of mass og flytt snoren gradvis
bakover til du får ustabil oppførsel fra raketten for å se hvor god
margin du har. Til slutt må det gjøres en avsluttende inspeksjon. Dette
inkluderer alle deler av oppskytingssystemet, raketten payload,

Sikker oppskyting av modellraketter

Denne sjekklisten er utviklet av Andøya Space Education med utgangspunkt
i Luftfartstilsynets forskrift om luftfartøy som ikke har fører ombord
og The National Association of Rocketry (NAR) sine retningslinjer for
sikker oppskyting av modellraketter. Denne veiledningen er laget for
elever, lærere, vitensenterpedagoger og andre modellrakettentusiaster
som en retningslinje for trygg bruk av modellraketter.

Generelle bestemmelser fra forskrift om luftfartøy som ikke har fører om
bord

Modellraketter faller under definisjonen modellfly som vi finner under
§4 Definisjoner og forkortelser i forskrift om luftfartøy som ikke har
fører om bord. Derfor forholder vi oss til de generelle bestemmelsene i
denne forskriften som gjelder modellfly.

«Modellfly: innretninger som ikke har fører om bord, som beveger seg i
luften og hvor flygingen kun har rekreasjon, sport eller konkurranse som
formål» – §4 Definisjoner og forkortelser, Forskrift om luftfartøy som
ikke har fører om bord.

Ifølge §6 Begrensninger for flygningen skal all flyging med modellfly
skje på en hensynsfull måte som ikke utsetter luftfartøy, personer,
fugler, dyr eller eiendom for risiko for skade eller for øvrig er til
sjenanse for allmennheten.

Modellflyet må til enhver tid være godt synlig for den som fører det
slik at full kontroll over modellflyet kan opprettholdes. Flyging med
modellfly kan bare skje i dagslysperioden og ikke:

1.  høyere enn 120 meter over bakken eller vannet

2.  nærmere enn 150 meter fra personer, motorkjøretøy eller bygning som
    ikke er under fartøyførerens kontroll med unntak av under start og
    landing

3.  nærmere enn 5 km fra lufthavner

**Sjekkliste for oppskyting av modellraketter**

1.  Materialer: Modellrakettens nesekone, kropp og finner skal bygges i
    lette materialer. Ikke metall eller ander tyngre materialer.

2.  Motorer: Det skal kun brukes sertifiserte motorer fra seriøse
    leverandører. Les alltid medfølgende bruksanvisning nøye før bruk av
    rakettmotorer.

3.  Antenning: Modellrakettene skal alltid skytes opp med en elektrisk
    tenner og elektrisk lunte. Tenneren skal ha en sikringsfunksjon som
    kan slås av og på. Avfyringsknappen skal av seg selv gå tilbake i av
    posisjon etter at den er trykket inn.

4.  Feiltenning: Hvis raketten ikke antennes skal sikring til tenneren
    slås på, eller batteriet til tenneren fjernes. Deretter skal en
    vente i minst 60 sekunder før man nærmer seg raketten.

5.  Sikkerhet ved oppskyting: Enhver oppskyting skal ha egen nedtelling
    slik at alle tilskuere er oppmerksomme på oppskytingen. Alle
    tilskuere skal holde minst 15 m avstand. Den som trykker på tenneren
    skal bruke vernebriller, og ha en avstand på minst 5 m fra
    utskytingsplattformen.

6.  Utskytingsrampe: Det skal brukes en dedikert rakettrampe, f.eks. en
    utskytningsplattform med startstang. Vinkelen til rakettrampen skal
    kunne justeres, og ved oppskyting skal vinkelen alltid være mellom
    70° og 85° fra horisontalplanet, slik at raketten flyr omtrent rett
    opp. Det skal også brukes en varmedeflektor for å forhindre den
    varme gassen fra motoren i å treffe bakken. Ved bruk av startstang
    bør enden på stanga sikres mellom hver oppskyting for å forhindre
    øyeskader.

7.  Størrelse: Rakettens startvekt skal ikke overgå 250 gram. Største
    totalimpuls på motoren skal ikke overgå 20 Ns (motorkategori D).

8.  Sikkerhet under flygning: Modellraketter skal kun skytes i klar
    sikt. Raketter skal ikke skytes ved lavt skydekke eller i nærheten
    av fugler, fly eller andre objekter i lufta. Ved bruk av nyttelast
    skal ikke denne være brennbar, eksplosiv eller utgjøre noen annen
    fare for raketten eller omgivelsene.

9.  Egnet sted for oppskyting: Raketter skal kun skytes utendørs i et
    stort åpent område med minst 200×200 m klaring, og hvor alle
    personer som oppholder seg innenfor dette området er informert om
    aktiviteten. Sjekk alltid værmeldingen før planlagt oppskyting av
    modellraketter. Rakettoppskytinger skal ikke skje ved middelvind som
    overgår 7 m/s. Oppskyting skal heller ikke skje i områder som er
    eksponert for skogbrann.

10. Fallskjerm: Enhver modellrakett skal ha fallskjerm slik at den kan
    lande sikkert. Det skal også alltid brukes et flammebestandig
    materiale inni raketten for å beskytte fallskjermens funksjon når
    denne skytes ut, f.eks. wadding.

11. Sikkerhet ved henting: Raketter skal ikke hentes hvis de lander på
    utilgjengelige steder, f.eks. høyt oppe i trær eller i
    strømledninger.
